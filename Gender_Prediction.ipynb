```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1 =cv2.imread('/content/selena-gomez-drake.jpg')
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
result=DeepFace.analyze(img1, actions=['gender'])
```

```python
print('Gender : ', result['gender'])
```
